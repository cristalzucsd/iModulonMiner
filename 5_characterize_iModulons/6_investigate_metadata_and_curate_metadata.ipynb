```python
# imports and loading
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['svg.fonttype'] = 'none'

metadata_df = pd.read_csv('../Zymomonas_mobilis/Zymomonas_mobilis.txt', delimiter = '\t', index_col = 0)
```

```python
# pie charting
meta_dict = dict(metadata_df.CenterName.value_counts())
meta_names = meta_dict.keys()
meta_vals = meta_dict.values()
plt.pie(meta_vals, labels = meta_names, autopct=lambda p: f'{p:.1f}% ({int(p * sum(meta_vals) / 100)})')
plt.savefig('pie_1.pdf', transparent = True)
plt.show()
```

```python
# pie charting
meta_dict = dict(metadata_df.ScientificName.value_counts())
meta_names = meta_dict.keys()
meta_vals = meta_dict.values()
plt.pie(meta_vals, labels = meta_names, autopct=lambda p: f'{p:.1f}% ({int(p * sum(meta_vals) / 100)})')
plt.savefig('pie_2.pdf', transparent = True)
plt.show()
```

```python
# let's add some more columns to generate a good metadata file for later use in iModulon characterization
new_to_old = {
    'sample_id' : '',
    'study' : '',
    'project' : '',
    'condition' : '',
    'rep_id' : '',
    'Strain Description' : 'ScientificName',
    'Strain' : 'TaxID',
    'Evolved Sample' : '',
    'Base Media' : '',
    'Temperature (C)' : '',
    'pH' : '',
    'Carbon Source (g/L)' : '',
    'Nitrogen Source (g/L)' : '',
    'Electron Acceptor' : '',
    'Supplement' : '',
    'Growth Rate (1/hr)' : '',
    'Isolate Type' : 'LibrarySelection',
    'Additional Details' : '',
    'project_reference' : '',
    'Sequencing Machine' : 'Model',
    'LibraryLayout' : 'LibraryLayout',
    'Platform' : 'Platform',
    'Biological Replicates' : '',
    'doi' : '',
    'GEO' : '',
    'SRX' : 'Experiment',
    'Run' : 'Run',
    'BioProject' : 'BioProject',
    'run_date' : 'ReleaseDate',
}
iM_metadata = pd.DataFrame(columns = new_to_old.keys(), index = metadata_df.index)
for new, old in new_to_old.items():
    if old != '':
        if old == 'Experiment':
            iM_metadata[new] = metadata_df.index
        else:
            iM_metadata[new] = metadata_df[old]
            
iM_metadata.to_csv('raw_metadata_file_for_iMs.csv')
```

```python
# already released plot
metadata_df['convert_release'] = pd.to_datetime(metadata_df['ReleaseDate'])

# Step 3: Count the number of releases per date
release_counts = metadata_df.groupby('convert_release').size()

# Step 4: Calculate the cumulative sum to get accumulated releases
cumulative_releases = release_counts.cumsum()

# Step 5: Plot the accumulated results
cumulative_releases.plot(kind='line',  title='Zymomonas Mobilis Samples')
plt.xlabel('Year')
plt.ylabel('Number of Samples')
plt.savefig('zym_num_samples.pdf', transparent = True)
plt.show()
```
